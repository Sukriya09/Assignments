Task- 3- Outlier Detection:
 Tasks:
 1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

 2. Define outliers using the 1.5 × IQR rule.

```python
#Outliers detection of Score
#calculate the first and third quartile
Q1_score = filtered_df['score'].quantile(0.25)  #Lower quartile(q1) is the 25th percentile of the data
Q3_score = filtered_df['score'].quantile(0.75)  #Upper quartile(q3) is the 75th percentile of the data

#calculate the iqr(interquartile range)
IQR_score = Q3_score = Q1_score

#Finding the lower bound and upper bound
lower_bound_score = Q1_score - 1.5 * IQR_score   #contains min value of range in a data
upper_bound_score = Q3_score + 1.5 * IQR_score  #contains max value

#identify the outlier countries based on the score
outlier_score = filtered_df[(filtered_df['score'] < lower_bound_score) | (filtered_df['score'] > upper_bound_score)]


```

```python
#outlier detection of Log GDP per capita
q1_gdp = filtered_df['Log GDP per capita'].quantile(0.25)
q3_gdp = filtered_df['Log GDP per capita'].quantile(0.75)

#calculate the IQR
IQR_gdp = q3_gdp = q1_gdp

#calculate upper and lower bound
lb_gdp = q1_gdp  - 1.5 * IQR_gdp
ub_gdp = q3_gdp + 1.5 * IQR_gdp

#identify outlier countries based on 'Log GDP per capita'
outlier_gdp = filtered_df[(filtered_df['Log GDP per capita'] < lb_gdp) | (filtered_df['Log GDP per capita'] > ub_gdp)]
```

```python
print("Outlier countries based on Score: \n")
outlier_score[['score','Country name']]
```

```python
print("Outlier countries based on Log GDP per capita: \n")
outlier_gdp[['Log GDP per capita','Country name']]
```
